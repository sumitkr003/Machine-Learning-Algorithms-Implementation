```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,1.0,1.0])
axes1.plot(x,y,'r')
axes1.set_xlabel('X1-axis')
axes1.set_ylabel('Y1-axis')
axes1.set_title('Axes 1')

axes2 = fig.add_axes([0.3,0.5,0.5,0.5])
axes2.plot(x,y,'b')
axes2.set_xlabel('X2-axis')
axes2.set_ylabel('Y2-axis')
axes2.set_title('Axes 2')
```

### Subplots

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
```

```python
fig
```

```python
axes
```

##### Here axes is just an array of objects of subplots.
therefore we can do something like this

```python
fig,axes = plt.subplots(1,2)

axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
#we can also iterate through axes array
fig,axes = plt.subplots(1,2)

for ax in axes:
    ax.plot(x,y)
    ax.set_xlabel('x-axis')
    ax.set_ylabel('y-axis')

plt.tight_layout() #to avoid overlapping of two plots
```

### Legends,Labels and titles

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label="x**2")
ax.plot(x, x**3, label="x**3")
ax.legend(loc=0) #loc=0 will best fit the legend in the plot and also by default it is 0
```
