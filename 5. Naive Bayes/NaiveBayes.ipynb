```python
import numpy as np
import pandas as pd

from sklearn.preprocessing import LabelEncoder
```

```python
dataset = pd.read_csv('./mushrooms.csv')
print(dataset.shape)
dataset.head()
```

```python
le = LabelEncoder()
df = dataset.apply(le.fit_transform)
df.head()
```

```python
split_factor = 0.90
split = int(df.shape[0] * split_factor)
print(split)
```

```python
train = df.iloc[:split]
test = df.iloc[split:]
print(train.shape)
print(test.shape)
```

```python
def prior_prob(data,label_column,label_value):
    n_rows = data[data[label_column] == label_value]
    return float(n_rows.shape[0]/data.shape[0])

prior_prob(df,'type',1)
```

```python
def cond_prob(data,feature_column,feature_value,label_column,label_value):
    n_rows = data.loc[data[label_column] == label_value]
    constraint_rows = data.loc[(data[label_column] == label_value) & (data[feature_column] == feature_value)]
    
    cp = constraint_rows.shape[0]/float(n_rows.shape[0])
    return cp

cond_prob(df,'cap_shape',3,'type',0)
```

```python
temp = df.values[:,0]
print(temp.shape)

x = np.unique(temp,return_counts=True)
print(x)
print(x[1][0]/float(x[1].sum()))
print(x[1][1]/float(x[1].sum()))
```

```python
#Naive Bayes algorithm implementation
CLASSES = np.unique(df['type'])
print(CLASSES)
print(type(CLASSES))
preds = []
ctr = 0

for ix in range(test.shape[0]):
    if ix%100 == 0:
        print(ctr)
    
    ctr+=1
    
    val = test.iloc[ix][1:]
    features = dict(val)
    
    posterior_prob = []
    for cx in CLASSES:
        likelihood = 1.0
        for fx in features:
            cp = cond_prob(train,fx,features[fx],'type',cx)
            likelihood*= cp
            
        prior = prior_prob(train,'type',cx)
        post_prob = likelihood * prior
        posterior_prob.append(post_prob)
        
    posterior_prob = np.array(posterior_prob)
    pred_label = posterior_prob.argmax()
    preds.append(pred_label)
```

```python
true_labels = np.array(test['type'])
print(true_labels.shape)

pred_labels = np.array(preds)
print(pred_labels.shape)

accuracy = np.sum(np.array(true_labels == pred_labels))/float(true_labels.shape[0])
print(accuracy)
```
