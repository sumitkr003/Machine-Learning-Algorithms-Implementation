```python
import pandas as pd
```

```python
df = pd.read_csv('./datasets/cal_housing_clean.csv')
```

```python
df.head()
```

```python
df.describe().transpose()
```

```python
X = df.drop('medianHouseValue',axis=1)
```

```python
X.head()
```

```python
Y = df['medianHouseValue']
Y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.3,random_state=0)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train = pd.DataFrame(data=scaler.transform(x_train),columns=x_train.columns,index=x_train.index)

#alternative way without minmaxscaler
### x_train = (x_train - x_train.min())/(x_train.max() - x_train.min())
```

```python
x_test = pd.DataFrame(data=scaler.transform(x_test),columns = x_test.columns,index=x_test.index)
```

```python
df.columns
```

```python
import tensorflow as tf
```

```python
age = tf.feature_column.numeric_column('housingMedianAge')
rooms = tf.feature_column.numeric_column('totalRooms')
bedrooms = tf.feature_column.numeric_column('totalBedrooms')
population = tf.feature_column.numeric_column('population')
household = tf.feature_column.numeric_column('households')
income = tf.feature_column.numeric_column('medianIncome')
```

```python
feat_cols = [age,rooms,bedrooms,population,household,income]
```

```python
input_fn = tf.estimator.inputs.pandas_input_fn(x=x_train,y=y_train,batch_size=10,num_epochs=1000,shuffle=True)
```

```python
model = tf.estimator.DNNRegressor(hidden_units=[6,6,6],feature_columns=feat_cols)
```

```python
model.train(input_fn=input_fn,steps=25000)
```

```python
test_input_fn = tf.estimator.inputs.pandas_input_fn(x=x_test,y=y_test,batch_size=10,num_epochs=1,shuffle=False)
```

```python
pred_gen = model.predict(test_input_fn)
```

```python
predictions = list(pred_gen)
```

```python
final_pred = []

for pred in predictions:
    final_pred.append(pred['predictions'][0])
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test,final_pred) ** 0.5
```
