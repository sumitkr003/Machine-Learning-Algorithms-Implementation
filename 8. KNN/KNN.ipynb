## KNN FROM SCRATCH

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Input data generation
mean_01 = np.array([0.0,0.0])
cov_01 = np.array([[1.0,0.1],[0.1,1.1]])
print(mean_01)
print(cov_01)
```

```python
mean_02 = np.array([5.0, 6.0])
cov_02 = np.array([[1.3, 0.2], [0.2, 1.5]])
```

```python
dist_01 = np.random.multivariate_normal(mean_01, cov_01, 500)
dist_02 = np.random.multivariate_normal(mean_02, cov_02, 500)

print(dist_01)
print(dist_02)
```

```python
for i in range(dist_01.shape[0]):
    plt.scatter(dist_01[i,0],dist_01[i,1],c='r')
    plt.scatter(dist_02[i,0],dist_02[i,1],c='b')
plt.show()
```

```python
rows = dist_01.shape[0] + dist_02.shape[0]
cols = dist_01.shape[1] + 1

data = np.zeros((rows,cols))

print(data.shape)
print(data.dtype)
```

```python
data[:dist_01.shape[0],:-1] = dist_01
data[dist_01.shape[0]:rows,:-1] = dist_02

data[dist_01.shape[0]:rows,-1] = 1.0
print(data.shape)
```

```python
np.random.shuffle(data)
print(data[:10])
```

```python
#Testing training split
split_factor = 0.85
split = int(split_factor * data.shape[0])
train_data = data[:split]
test_data = data[split:]
print (train_data.shape)
print (test_data.shape)

X_train = train_data[:,:-1]
y_train = train_data[:,-1]

X_test = test_data[:,:-1]
y_test = test_data[:,-1]

print (X_train.shape)
print (X_test.shape)
```

```python
print(np.unique(y_train,return_counts=True))
print(np.unique(y_test))      
```

```python
#Eucledian distance
def distance(p1,p2):
    dist = np.sqrt(((p1-p2)**2).sum())
    return dist

#Manhattan_distance
def distance_manhat(p1,p2):
    return np.sum(abs(p1-p2))

print (distance(np.array([0.0,0.0]),np.array([1.0,1.0])))
print (distance_manhat(np.array([0.0,0.0]),np.array([1.0,1.0])))
```

```python
#This function will return predicted label for a test point
def KNN(test_point,x_train,y_train,k=5):
    d=[]
    
    for i in range(x_train.shape[0]):
        dist = distance(test_point,x_train[i])
        d.append((dist,y_train[i]))
    
    d = np.array(d)
    sorted_vals = sorted(d,key=lambda z:z[0])
    sorted_vals = np.array(sorted_vals)[:k]
    
    freq = np.unique(sorted_vals[:,:],return_counts=True)
    label = freq[0][freq[1].argmax()]
    
    return label
```

```python
test_point = np.array([3.0,2.0])
print (KNN(test_point,X_train,y_train))
```

```python
def get_accuracy(x_train,y_train,x_test,y_test):
    y_pred = []
    for i in range(x_test.shape[0]):
        y_pred.append(KNN(x_test[i],x_train,y_train))
    
    y_pred = np.array(y_pred)
    ans = np.array(y_pred == y_test)
    acc = np.sum(ans)/float(x_test.shape[0])
    
    return acc
```

```python
print (get_accuracy(X_train,y_train,X_test,y_test))
```

## KNN Using Scikit Learn

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```
